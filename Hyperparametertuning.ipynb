# 선형회귀를 이용하여 학습하고, 하이퍼패러미터를 변경시켜가며 교차검증 후 가장 좋은 성능을 갖는 하이퍼패러미터를 찾아보시오.
시멘트의 강도에 대해 탐구,
strength? 에 대한 부분 예측

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ralbu85/DataScience_2022S/master/data/concrete.csv')
df # data set
```

```python
y = df['strength']
X = df.drop(columns=['strength']) ## 전체 칼럼에서 strength 제외
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_validate
from sklearn.linear_model import Ridge, Lasso
import warnings
from sklearn.preprocessing import PolynomialFeatures
warnings.filterwarnings('ignore')
```

test_neg_mean_squared의 절대값이 작으면 작을수록 오차가 작다. 하이퍼파라미터 람다를 0.1~100 모든 조합을 다 비교해보면서 정확도가 더 좋아지는 조합이 있는지 확인 해볼것이다.

```python
kfold = KFold(5,shuffle=True)
for i in range(1,8):
    for j in [0,0.001,0.01,0.05,0.1,0.15,0.2,0.25,0.3]:
        transformer = PolynomialFeatures(degree=i, interaction_only=False)
        X_=transformer.fit_transform(X)
        reg= Lasso(alpha=j)    
        result=cross_validate(estimator=reg, X=X_, y=y, cv=kfold, scoring=['neg_mean_squared_error','r2'], return_train_score=True)
        print(i,j,'test',result['test_neg_mean_squared_error'].mean(), result['test_r2'].mean())
```

작은 절대값을 가진 차수 라인은 test_neg_mean_squared 는 k가 4~5값을 가질때이다.
k가 4와 5일 때의 k와 lambda의 조합을 조금 더 찾아보도록 하겠다.

```python
for i in range(4,6):
    for j in [0,0.001,0.005,0.01,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5]:
        transformer = PolynomialFeatures(degree=i, interaction_only=False)
        X_=transformer.fit_transform(X)
        reg= Lasso(alpha=j)    
        result=cross_validate(estimator=reg, X=X_, y=y, cv=kfold, scoring=['neg_mean_squared_error','r2'], return_train_score=True)
        print(i,j,'test',result['test_neg_mean_squared_error'].mean(), result['test_r2'].mean())
```
